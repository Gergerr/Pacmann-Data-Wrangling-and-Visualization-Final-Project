# 31. Answering the first objective
What is our first objective?

1. Identifying High-Risk Customers: Those who frequently take cash-advance or have high balance.

```python
data['HIGH_RISK_CASH_ADVANCE'] = pd.cut(data['CASH_ADVANCE_FREQUENCY'], bins=[0, 0.1, 0.5, data['CASH_ADVANCE_FREQUENCY'].max()], labels=['Low', 'Medium', 'High'])
data['HIGH_RISK_BALANCE'] = pd.cut(data['BALANCE'], bins=[0, 1000, 3000, data['BALANCE'].max()], labels=['Low', 'Medium', 'High'])

# Grouping data to see average cash advance and balance in each category
risk_group = data.groupby(['HIGH_RISK_CASH_ADVANCE', 'HIGH_RISK_BALANCE']).agg({
    'CASH_ADVANCE': 'mean',
    'BALANCE': 'mean',
    'CREDIT_LIMIT': 'mean',
    'PAYMENTS': 'mean'
}).reset_index()

risk_group.head()
```

The reason I chose this is to categorize and analyze customer data based on two key indicators of financial risk: cash advance frequency and account balance. The intention is to identify patterns or clusters of behavior that could signify higher or lower financial risk.

Why?
* Cash Advance Frequency: Frequent cash advances can indicate reliance on credit for daily liquidity, which may be a sign of underlying financial instability or distress. By categorizing this frequency into 'Low', 'Medium', and 'High', we can differentiate between occasional and potentially risky habitual reliance on cash advances.

* Balance: High balances, especially those close to or at the credit limit, can also signal potential financial stress or risk. Segmenting balances into categories helps in identifying customers who may be over-leveraged.

```python
# Correctly creating a pivot table for the heatmap
risk_heatmap_data = data.pivot_table(index='HIGH_RISK_CASH_ADVANCE', columns='HIGH_RISK_BALANCE', values='BALANCE', aggfunc='mean')

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(risk_heatmap_data, annot=True, fmt=".0f", cmap='Reds', cbar_kws={'label': 'Average Balance'})
plt.title('Risk Levels by Cash Advance Frequency and Balance')
plt.xlabel('Balance Risk Level')
plt.ylabel('Cash Advance Risk Level')
plt.show()
```

```python
# Assuming weights for each risk factor based on hypothetical business insights
data['RISK_SCORE'] = (
    0.5 * pd.cut(data['BALANCE'], bins=3, labels=False) +
    0.3 * pd.cut(data['CASH_ADVANCE_FREQUENCY'], bins=3, labels=False) +
    0.2 * (1 - pd.cut(data['PRC_FULL_PAYMENT'], bins=3, labels=False))
)

# Visualizing the distribution of the risk score
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(data['RISK_SCORE'], bins=10, kde=False)
plt.title('Distribution of Composite Risk Score')
plt.xlabel('Risk Score')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Example: Chi-Square Test between risk category and a categorical outcome like delinquency status
from scipy.stats import chi2_contingency

# Hypothetical coding of delinquency status
data['DELINQUENCY_STATUS'] = (data['BALANCE'] > 3000).astype(int)  # Simplified assumption

contingency_table = pd.crosstab(data['BALANCE_CATEGORY'], data['DELINQUENCY_STATUS'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square test result -- chi2: {chi2}, p-value: {p}")
```

```python
# Grouping data by 'BALANCE_CATEGORY' and calculating average values of other risk factors
grouped_data = data.groupby('BALANCE_CATEGORY').agg({
    'CASH_ADVANCE_FREQUENCY': 'mean',
    'PRC_FULL_PAYMENT': 'mean',
    'CREDIT_LIMIT': 'mean',
    'PAYMENTS': 'mean'
}).reset_index()

print(grouped_data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar plot for each factor by balance category
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

sns.barplot(x='BALANCE_CATEGORY', y='CASH_ADVANCE_FREQUENCY', data=grouped_data, ax=axs[0, 0])
axs[0, 0].set_title('Average Cash Advance Frequency by Balance Category')

sns.barplot(x='BALANCE_CATEGORY', y='PRC_FULL_PAYMENT', data=grouped_data, ax=axs[0, 1])
axs[0, 1].set_title('Average Full Payment Rate by Balance Category')

sns.barplot(x='BALANCE_CATEGORY', y='CREDIT_LIMIT', data=grouped_data, ax=axs[1, 0])
axs[1, 0].set_title('Average Credit Limit by Balance Category')

sns.barplot(x='BALANCE_CATEGORY', y='PAYMENTS', data=grouped_data, ax=axs[1, 1])
axs[1, 1].set_title('Average Payments by Balance Category')

plt.tight_layout()
plt.show()
```

# 32. Answering the second objective

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Analyzing distributions of One-off and Installment Purchases
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

sns.histplot(data['ONEOFF_PURCHASES'], bins=30, ax=axs[0], kde=True, color='skyblue')
axs[0].set_title('Distribution of One-off Purchases')
axs[0].set_xlabel('One-off Purchases')
axs[0].set_ylabel('Frequency')

sns.histplot(data['INSTALLMENTS_PURCHASES'], bins=30, ax=axs[1], kde=True, color='lightgreen')
axs[1].set_title('Distribution of Installment Purchases')
axs[1].set_xlabel('Installment Purchases')
axs[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
data['INSTALLMENTS_PURCHASES'].describe(), data['ONEOFF_PURCHASES'].describe()
```

After a thoughtful consideration and guesses,
I set the following category:

|ONEOFF_PURCHASES|SPENDING_TYPE|
| :---: | :---: |
| $\le$ 1| NONE |
| $\le$ 100 | LOW |
| $\le$ 750 | MODERATE |
| $>$ 750 | HIGH |

and

|INSTALLMENT_PURCHASES|INSTALLMENT_TYPE|
| :---: | :---: |
| $\le$ 1| NONE |
| $\le$ 100 | LOW |
| $\le$ 600 | MODERATE |
| $>$ 600 | HIGH |

```python
import numpy as np

# Define custom bins based on the distribution observed from data['ONEOFF_PURCHASES'].describe() and similar for installments
oneoff_edges = [-np.inf, 1, 100, 750, np.inf]  # Assuming most values are 0, set the first bin edge between 0 and 1
installment_edges = [-np.inf, 1, 100, 600, np.inf]  # Adjust accordingly based on actual data

# Categorizing customers based on custom bins
data['SPENDING_TYPE'] = pd.cut(data['ONEOFF_PURCHASES'], bins=oneoff_edges, labels=['None', 'Low', 'Moderate', 'High'])
data['INSTALLMENT_TYPE'] = pd.cut(data['INSTALLMENTS_PURCHASES'], bins=installment_edges, labels=['None', 'Low', 'Moderate', 'High'])

# Viewing the new categorizations
print(data[['ONEOFF_PURCHASES', 'SPENDING_TYPE', 'INSTALLMENTS_PURCHASES', 'INSTALLMENT_TYPE']].head())
```

```python
# Creating credit utilization ratio
data['UTILIZATION_RATIO'] = data['BALANCE'] / data['CREDIT_LIMIT']

# Segmenting by utilization
data['UTILIZATION_TYPE'] = pd.qcut(data['UTILIZATION_RATIO'], 4, labels=['Low', 'Moderate', 'High', 'Very High'])

# Plotting the distribution of Credit Utilization
plt.figure(figsize=(10, 6))
sns.histplot(data['UTILIZATION_RATIO'], bins=30, kde=True, color='purple')
plt.title('Distribution of Credit Utilization Ratios')
plt.xlabel('Credit Utilization Ratio')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Aggregate data by Spending and Utilization Type for visualization
segment_analysis = data.groupby(['SPENDING_TYPE', 'UTILIZATION_TYPE']).agg({
    'BALANCE': 'mean',
    'PAYMENTS': 'mean',
    'CREDIT_LIMIT': 'mean',
    'CUST_ID': 'count'
}).rename(columns={'CUST_ID': 'COUNT'}).reset_index()

# Visualizing customer segments
sns.catplot(x='UTILIZATION_TYPE', y='BALANCE', hue='SPENDING_TYPE', data=segment_analysis, kind='bar', height=5, aspect=2)
plt.title('Customer Segments by Spending and Utilization Type')
plt.ylabel('Average Balance')
plt.xlabel('Utilization Type')
plt.show()
```
